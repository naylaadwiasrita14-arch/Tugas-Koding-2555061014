# Matrices (Rank 2 Tensors) in NumPy

```python
import numpy as np
```

```python
# buat matrix X berukuran 2x3 menggunakan numpy
x = np.array ([[21,45,8],
                   [10,5, 90]])
x
```

```python
# tampilkan ukuran dari matrix X
x. shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x [:,0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x [1, :]
```

# Matrices in PyTorch

```python
import torch
```

```python
# buat matrix X berukuran 2x3 menggunakan pytorch
x_pt = torch.tensor([[21,45,8],
                   [10,5, 90]])
x_pt
```

```python
# tampilkan ukuran dari matrix X
x_pt.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x_pt[0, :]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x_pt[1, :]
```

# *Matrices in TensorFlow*

```python
import tensorflow as tf
```

```python
# buat matrix X berukuran 2x3 menggunakan TensorFlow
x_tf = tf.Variable([[21,45,8],
                    [10,5,90]])
x_tf
```

```python
# tampilkan ukuran dari matrix X
x_tf.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x_tf[0, :]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x_tf[1, :]
```
